<h3>Data Cleaning</h3>
The process of fixing or removing incomplete, irrelevent and incompetent Data.
75% of work done in pandas is Data cleaning.

Real-world Data is messy. Pandas makes it incredibly easy to handle missing data(NaN), 
drop duplicate-rows, filter-out irrelvent information and rename columns.

Here are some of the few methods for Data-cleaning.

```python
"""
Suppose you want to drop an irrelevent column.For this purpose, use the
df.drop() method. You can also pass the list of multiple columns that you
want to drop separated by comma.
"""


df = df.drop(columns=["OrderDate"])
print(df.head())
```

```python
"""
Suppose you have NaN values in your DataFrame. To drop them, We will Use df.dropna() method.
It will drop the entire row that contains NaN.
"""

"""I have modified my .csv file by typing NaN values to some of the rows.."""

df_new = df.dropna()

print(df_new.head())


# As you can see in the output that the row at index 1 is dropped due to containing NaN value.
```

```python
"""
Suppose that you have NaN value in Your DataFrames and Instead of Dropping the entire 
row, You want that NaN value to be filled with some value. For This purpose, We will Use 
df.fillna() method.
"""

print(df.head())

# As you can see in the output, Our DataFrame contains multiple Nans.
```

```python
df = df.fillna({"Price" : 0, "Quantity" : 0})
print(df.head())

# We replaced the NaN values in "Price" and "Quantity" columns with 0.0
```

```python
"""
Suppose we have a DataFrame that contains duplicate rows and we Want to drop them.
For that purpose we will use the df.drop_duplicates method..

"""

df = pd.read_csv("orders.csv")
print(df.head())

# As you can see in the output, we have duplicate rows.
```

<h1>Exercise 1:-</h1>

Scenario: You are a full-stack engineer looking at a corrupted transaction log from a Flask backend. The data has duplicates, missing price metrics, and unstandardized text.

```python
# Here is your raw DataFrame setup code:

import pandas as pd
import numpy as np

raw_data = {
    'TransactionID': [101, 102, 103, 104, 104, 105, 106, 107],
    'Customer': ['ameer', 'Sarah', 'ameer', 'Zayn', 'Zayn', 'Sarah', 'ameer', 'Bilal'],
    'Category': ['Electronics', 'Books', 'electronics', 'Clothing', 'Clothing', 'Books', 'Clothing', 'Books'],
    'Price': [1200.0, 25.0, 450.0, np.nan, np.nan, 150.0, 120.0, 45.0],
    'Quantity': [1, 2, 1, 3, 3, 1, 2, np.nan]
}

df = pd.DataFrame(raw_data)
print(df)
```

```python
# Task 1: remove the completly duplicates rows based on transactionID.

"""To remove the duplicates, we will use the .drop_duplicates() method"""

df = df.drop_duplicates(subset=["TransactionID"], keep="first")

# Task 2: The Customer names and Category values have inconsistent casing 
# (e.g., 'ameer' vs 'Sarah', 'Electronics' vs 'electronics'). Force all text columns to be strictly lowercase.

df["Customer"] = df["Customer"].str.lower()
df["Category"] = df["Category"].str.lower()

# Task 3: Fill missing values in Price with the median price of the data. 
# Fill missing values in Quantity with a default value of 1.


df = df.fillna({"Price": df["Price"].median(), "Quantity" : 1})


# Task 4:  Create a new column called TotalCost which is calculated as $Price \times Quantity$.

df["TotalCost"] = df["Price"] * df["Quantity"]


print(df)
```

```python
saas_data = {
    'ClientID': ['C_101', 'C_102', 'C_103', 'C_104', 'C_105', 'C_106'],
    'Region': ['PK', 'AE', 'PK', 'SA', 'AE', 'UK'],
    'MonthlySpend': [2500, 150, 4500, 80, 950, 12000],
    'Status': ['Active', 'Churned', 'Active', 'Pending', 'Active', 'Active']
}

df_saas = pd.DataFrame(saas_data)
df_saas.set_index("ClientID", inplace=True)

print(df_saas)
```

<h2>Exercise 3:</h2>

The Split-Apply-Combine Engine:

Here is your raw dataset representing developer commits across a decentralized tech team:

```python
dev_data = {
    'Repo': ['Backend', 'Frontend', 'Backend', 'DevOps', 'Frontend', 'Backend', 'DevOps'],
    'Language': ['Python', 'React', 'Python', 'Go', 'React', 'Go', 'Go'],
    'Commits': [12, 5, 22, 8, np.nan, 14, 30],
    'LinesOfCode': [450, 120, 890, 300, 50, 410, np.nan]
}

df_dev = pd.DataFrame(dev_data)

print(df_dev)
```

```python
# Task 1: Fill the missing values in Commits with 0, and fill the missing values in LinesOfCode with 
# the mean value of the LinesOfCode column. Do this directly on df_dev

df_dev = df_dev.fillna({"Commits" : 0, "LinesOfCode" : df_dev["LinesOfCode"].mean()})
print(df_dev)
print()

# Task 2: Group the data by Repo. In a single operation, compute the total (sum) number 
# of Commits AND the average (mean) LinesOfCode for each repository.

repo_group = df_dev.groupby("Repo").agg({
    "Commits" : "sum",
    "LinesOfCode" : "mean"
})
print(repo_group)
print()

# task 3: Group the data by Language and filter out or extract the total sum of Commits 
# sorted in descending order (highest commits first).

lang_group = df_dev.groupby("Language")["Commits"].sum().sort_values(ascending=False)
print(lang_group)
```

<h1>Exercise 4: </h1>

you must merge two separate datasets and apply custom vectorized element-wise modifications 
using .map() or .apply().

Here is your raw production setup representing an e-commerce order tracking system:

```python
# Table A: Base Orders
orders_data = {
    'OrderID': [2001, 2002, 2003, 2004],
    'SKU': ['E_LPT', 'B_NOVEL', 'E_LPT', 'C_HOOD'],
    'Quantity': [1, 3, 2, 2],
    'Customer': ['Ameer', 'Sarah', 'Zayn', 'Bilal']
}

df_orders = pd.DataFrame(orders_data)
print("-------Orders Data------")
print(df_orders)
print()

# Table B: Product Catalog (Prices)
catalog_data = {
    'ProductSKU': ['E_LPT', 'B_NOVEL', 'C_HOOD'],
    'BasePrice': [1200, 20, 80],
    'TaxClass': ['High', 'Exempt', 'Standard']
}
df_catalog = pd.DataFrame(catalog_data)
print("------------Catalog Data---------------")
print(df_catalog)
print()
```

```python
# Task 1: Merge df_orders with df_catalog. You must map SKU from the orders table to ProductSKU 
# from the catalog table. Ensure your resulting DataFrame retains all rows from df_orders. 
# Dropping columns you don't need later is fine, but the merge must be mathematically exact.

df_merged = pd.merge(df_orders,df_catalog,left_on="SKU", right_on="ProductSKU",how="left")
print(df_merged)
print()
# Create a new column called TaxRate. Use a vectorized .map() or .apply() function using this 
# strict business logic dictionary:

tax_rules = {'High': 0.18, 'Standard': 0.05, 'Exempt': 0.00}

df_merged["TaxRate"] = df_merged["TaxClass"].map(tax_rules)
print("-----------------new data--------------------")
print(df_merged)


# Task 3: The Ultimate Financial Column: Compute a final column called FinalInvoiceAmount using the formula:
# $$FinalInvoiceAmount = (Quantity \times BasePrice) \times (1 + TaxRate)$$
# Print the final sorted DataFrame showing the highest invoice amounts first.

df_merged["FinalInvoiceAmount"] = (df_merged["Quantity"] * df_merged["BasePrice"]) * (1 + df_merged["TaxRate"])

print(df_merged.sort_values("FinalInvoiceAmount", ascending=False))
```
